```python
def get_pos(my_text):
    nlp_tokens = nlp(my_text)
    return [token.pos_ for token in nlp_tokens]

def get_tokens(my_text):
    nlp_tokens = nlp(my_text)
    return [token for token in nlp_tokens]

def get_key_pos(my_list):
    res = []
    for keyword in my_list:
        phrase = []
        for token in nlp(keyword):
            phrase.append(token.pos_)
        res.append(phrase)
    return res
```

# explaination of the code
So if a term repeats itself for more than once in a sequence we add a +

but in general regex terminology:
- we use `+` for one or more than one
- we use `*` for zero or more than one

So i am planning to use this if the current methods dont give results

```python
# Cocvert a list of POS tag sequences into a regular exp pattern
def pos_to_regex(key_pos_list):    
    regex_patterns = []
    
    for pos_sequence in key_pos_list:
        regex_part = []
        prev_pos = ''
        pos_count = 0
        for pos in pos_sequence:
            if pos == prev_pos:
                pos_count += 1
            else:
                if prev_pos:
                    regex_part.append(prev_pos + ('+' if pos_count > 1 else ''))
                prev_pos = pos
                pos_count = 1
        regex_part.append(prev_pos + ('+' if pos_count > 1 else ''))        
        regex_patterns.append(','.join(regex_part))
    
    return regex_patterns


key_pos_list = [['ADJ', 'NOUN', 'NOUN'], ['ADJ', 'NOUN', 'NOUN', 'ADJ', 'ADJ'], ['ADJ', 'NOUN', 'NOUN', 'ADJ', 'ADJ', 'ADJ'], ['ADJ','ADJ']]
regex_patterns = pos_to_regex(key_pos_list)
print(regex_patterns)
```

```python
def list_to_sorted_dict(my_list):
    counts = Counter(my_list)
    sorted_dict = dict(sorted(counts.items(), key=lambda item: item[1], reverse=True))
    return sorted_dict
```

```python
def categorize_and_analyze_keyphrases(key_pos_lists):
    categorized_keyphrases = {}
    for key_pos in key_pos_lists: 
        n_gram_length = len(key_pos)
        if categorized_keyphrases.get(n_gram_length):
            categorized_keyphrases[n_gram_length].append(key_pos)
        else:
            categorized_keyphrases[n_gram_length]=[key_pos]

    regex_freq_dict = {}
    for n_length, key_pos_list in categorized_keyphrases.items():
        regex_pattern_list = pos_to_regex(key_pos_list)
        regex_freq_dict[n_length] = list_to_sorted_dict(regex_pattern_list)

    regex_freq_dict = dict(sorted(regex_freq_dict.items(), key=lambda item: item[0], reverse=True))
    return categorized_keyphrases, regex_freq_dict
```
