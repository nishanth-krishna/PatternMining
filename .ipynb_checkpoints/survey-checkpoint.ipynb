```python
def calculate_f1_score(actual_keyphrases, predicted_keyphrases):
    actual_set = set([s.lower() for s in actual_keyphrases])
    predicted_set = set(s.lower() for s in predicted_keyphrases)

    true_positives = len(actual_set.intersection(predicted_set))
    precision = true_positives / len(predicted_set) if predicted_set else 0
    recall = true_positives / len(actual_set) if actual_set else 0

    if precision + recall == 0:
        f1_score = 0
    else:
        f1_score = 2 * (precision * recall) / (precision + recall)

    return f1_score


actual_keyphrases = ['keyphrase one', 'keyphrase two', 'keyphrase three']
predicted_keyphrases = ['keyphrase one', 'keyphrase four', 'keyphrase two']
f1_score = calculate_f1_score(actual_keyphrases, predicted_keyphrases)
print(f"F1 Score: {f1_score}")
```

```python
def get_hotwords(text):
    result = []
    pos_tag = ['PROPN', 'ADJ', 'NOUN'] 
    doc = nlp(text.lower()) 
    for token in doc:
        if(token.text in nlp.Defaults.stop_words or token.text in punctuation):
            continue
        if(token.pos_ in pos_tag):
            result.append(token.text)
    return result
new_text = """
When it comes to evaluating the performance of keyword extractors, you can use some of the standard metrics in machine learning: accuracy, precision, recall, and F1 score. However, these metrics don’t reflect partial matches. they only consider the perfect match between an extracted segment and the correct prediction for that tag.
Fortunately, there are some other metrics capable of capturing partial matches. An example of this is ROUGE.
"""
output = set(get_hotwords(new_text))
most_common_list = Counter(output).most_common(10)
for item in most_common_list:
    print(item[0])
```

```python
from collections import OrderedDict
import numpy as np
import spacy
from spacy.lang.en.stop_words import STOP_WORDS

nlp = spacy.load('en_core_web_sm')

class TextRank4Probality():
    """Extract keywords from text"""
    
    def __init__(self):
        self.d = 0.85 # damping coefficient, usually is .85
        self.min_diff = 1e-5 # convergence threshold
        self.steps = 10 # iteration steps
        self.node_weight = None # save keywords and its weight

    
    def set_stopwords(self, stopwords):  
        """Set stop words"""
        for word in STOP_WORDS.union(set(stopwords)):
            lexeme = nlp.vocab[word]
            lexeme.is_stop = True
    
    def sentence_segment(self, doc, candidate_pos, lower):
        """Store those words only in cadidate_pos"""
        sentences = []
        for sent in doc.sents:
            selected_words = []
            for token in sent:
                # Store words only with cadidate POS tag
                print(f"token:{token} \t POS:{token.pos_}")
                if token.pos_ in candidate_pos and token.is_stop is False:
                    if lower is True:
                        selected_words.append(token.text.lower())
                    else:
                        selected_words.append(token.text)
            sentences.append(selected_words)
        
        # print(f"SENTENCAES ARE: \n {sentences}")
        return sentences

        
    def get_vocab(self, sentences):
        """Get all tokens"""
        vocab = OrderedDict()
        i = 0
        for sentence in sentences:
            for word in sentence:
                if word not in vocab:
                    vocab[word] = i
                    i += 1
        return vocab
    
    def get_token_pairs(self, window_size, sentences):
        """Build token_pairs from windows in sentences"""
        token_pairs = list()
        for sentence in sentences:
            for i, word in enumerate(sentence):
                for j in range(i+1, i+window_size):
                    if j >= len(sentence):
                        break
                    pair = (word, sentence[j])
                    if pair not in token_pairs:
                        token_pairs.append(pair)
        return token_pairs
        
    def symmetrize(self, a):
        return a + a.T - np.diag(a.diagonal())
    
    def get_matrix(self, vocab, token_pairs):
        """Get normalized matrix"""
        # Build matrix
        vocab_size = len(vocab)
        g = np.zeros((vocab_size, vocab_size), dtype='float')
        for word1, word2 in token_pairs:
            i, j = vocab[word1], vocab[word2]
            g[i][j] = 1
            
        # Get Symmeric matrix
        g = self.symmetrize(g)
        
        # Normalize matrix by column
        norm = np.sum(g, axis=0)
        g_norm = np.divide(g, norm, where=norm!=0) # this is ignore the 0 element in norm
        
        return g_norm

    
    def get_keywords(self, doc, number=10):
        """Print top number keywords"""
        return_list = []
        node_weight = OrderedDict(sorted(self.node_weight.items(), key=lambda t: t[1], reverse=True))
        for i, (key, value) in enumerate(node_weight.items()):
            return_list.append(key)
            # print(key + ' - ' + str(value))
            if i > number:
                break

        return return_list
        # print("calling get_unigrams")
        # unigrams = return_list       # getting unigrams from the extracted keyphrases
        # print(f"finished get_unigrams: {unigrams}")
        # print("calling get_sequential_tokens")
        # sents = self.get_sequential_tokens(doc, lower=False)       # get sequential terms from text instaed of computing all the possible permutations of the unigrams
        # print(f"finished get_unigrams: {sents}")
        # print("calling get_n_grams")
        # n_grams = self.get_n_grams(sents)       # get the n_grams from interested POS tags
        # print(f"finished get_n_grams: {n_grams}")
        # final_key_phrases = self.find_longest_sequence(n_grams, unigrams)
        
        # return final_key_phrases
        
        
    def extract_key_phrases(self, text, 
                candidate_pos=['NOUN', 'PROPN', 'VERB', 'ADJ'], 
                window_size=4, lower=False, stopwords=list(), number=20):
        """Main function to analyze text"""
        
        # Set stop words
        self.set_stopwords(stopwords)
        
        # Pare text by spaCy
        doc = nlp(text)
        # print(doc)
        # print("---"*20)
        
        # Filter sentences
        sentences = self.sentence_segment(doc, candidate_pos, lower) # list of list of words
        # print(sentences)
        # print("---"*20)
        
        # Build vocabulary
        vocab = self.get_vocab(sentences)
        
        # Get token_pairs from windows
        token_pairs = self.get_token_pairs(window_size, sentences)
        
        # Get normalized matrix
        g = self.get_matrix(vocab, token_pairs)
        
        # Initionlization for weight(pagerank value)
        pr = np.array([1] * len(vocab))
        
        
        # Iteration
        previous_pr = 0
        for epoch in range(self.steps):
            pr = (1-self.d) + self.d * np.dot(g, pr)
            if abs(previous_pr - sum(pr))  < self.min_diff:
                break
            else:
                previous_pr = sum(pr)

        # Get weight for each node
        node_weight = dict()
        for word, index in vocab.items():
            node_weight[word] = pr[index]
        
        self.node_weight = node_weight

        print("calling get_keywords")
        final_key_phrases = self.get_keywords(doc, number) # get keyphrasess call
        print("finished get_keywords")

        return final_key_phrases

    def get_sequential_tokens(self, doc, lower):
        # to get sequential tokens from the text
        interested_pos = ['NOUN', 'PROPN', 'VERB', 'ADJ']
        ignore_pos = ['SCONJ', 'DET', 'ADP', 'AUX', 'PART', 'PUNCT', 'CCONJ', 'PRON', 'NUM', 'ADV']
        sentences = []
        for sent in doc.sents:
            selected_words = []
            for token in sent:
                if token.pos_ not in ignore_pos:
                    if lower:
                        selected_words.append(token.text.lower())
                    else:
                        selected_words.append(token.text)
                else:
                    selected_words.append("*")
            sentences.append(selected_words)
        return sentences
    
    def get_n_grams(self, sentences):
        # to get n_grams outta sequential tokens
        n_grams = []
        for sent in sentences:
            temp_list = []
            for word in sent:
                if word == '*':
                    if len(temp_list)!=0:
                        n_grams.append(temp_list)
                        temp_list = []
                else:
                    temp_list.append(word)
        return n_grams

    def find_longest_sequence(self, n_grams, unigrams):
        longest_sequences = []
        
        for n_gram in n_grams:
            current_sequence = []
            longest_sequence = []
            
            for word in n_gram:
                if word in unigrams:
                    current_sequence.append(word)
                    if len(current_sequence) > len(longest_sequence):
                        longest_sequence = current_sequence.copy()
                else:
                    current_sequence = []
            
            if len(longest_sequence)>0:
                longest_sequences.append(longest_sequence)
            
        longest_sequences = [' '.join(i) for i in sorted(longest_sequences, key=len, reverse=True)]
        
        return longest_sequences

    def find_longest_sequence(self, n_grams, unigrams):
        longest_sequences = []
        
        for n_gram in n_grams:
            current_sequence = []
            longest_sequence = []
            
            for word in n_gram:
                if word in unigrams:
                    current_sequence.append(word)
                    if len(current_sequence) > len(longest_sequence):
                        longest_sequence = current_sequence.copy()
                else:
                    current_sequence = []
            
            if len(longest_sequence)>0:
                longest_sequences.append(longest_sequence)
            
        longest_sequences = [' '.join(i) for i in sorted(longest_sequences, key=len, reverse=True)]
        
        return longest_sequences
```

```python
def remove_substrings(strings):
    # Sort strings by length in descending order to prioritize longer strings
    strings_sorted = sorted(strings, key=len, reverse=True)
    
    # Set to keep track of substrings to remove
    to_remove = set()
    
    # Compare each string with every other string
    for i, str_i in enumerate(strings_sorted):
        for j, str_j in enumerate(strings_sorted):
            if i != j and str_i in str_j:
                # If str_i is a substring of str_j, mark str_i for removal
                to_remove.add(str_i)
    
    # Filter out the substrings marked for removal
    result = [str_i for str_i in strings if str_i not in to_remove]
    return result
```

```python
from collections import OrderedDict
import numpy as np
import spacy
from spacy.lang.en.stop_words import STOP_WORDS

nlp = spacy.load('en_core_web_sm')

class TextRank4Keyword():
    """Extract keywords from text"""
    
    def __init__(self):
        self.d = 0.85 # damping coefficient, usually is .85
        self.min_diff = 1e-5 # convergence threshold
        self.steps = 10 # iteration steps
        self.node_weight = None # save keywords and its weight

    
    def set_stopwords(self, stopwords):  
        """Set stop words"""
        for word in STOP_WORDS.union(set(stopwords)):
            lexeme = nlp.vocab[word]
            lexeme.is_stop = True
    
    def sentence_segment(self, doc, candidate_pos, lower):
        """Store those words only in cadidate_pos"""
        sentences = []
        for sent in doc.sents:
            selected_words = []
            for token in sent:
                # Store words only with cadidate POS tag
                if token.pos_ in candidate_pos and token.is_stop is False:
                    if lower is True:
                        selected_words.append(token.text.lower())
                    else:
                        selected_words.append(token.text)
            sentences.append(selected_words)
        return sentences
        
    def get_vocab(self, sentences):
        """Get all tokens"""
        vocab = OrderedDict()
        i = 0
        for sentence in sentences:
            for word in sentence:
                if word not in vocab:
                    vocab[word] = i
                    i += 1
        return vocab
    
    def get_token_pairs(self, window_size, sentences):
        """Build token_pairs from windows in sentences"""
        token_pairs = list()
        for sentence in sentences:
            for i, word in enumerate(sentence):
                for j in range(i+1, i+window_size):
                    if j >= len(sentence):
                        break
                    pair = (word, sentence[j])
                    if pair not in token_pairs:
                        token_pairs.append(pair)
        return token_pairs
        
    def symmetrize(self, a):
        return a + a.T - np.diag(a.diagonal())
    
    def get_matrix(self, vocab, token_pairs):
        """Get normalized matrix"""
        # Build matrix
        vocab_size = len(vocab)
        g = np.zeros((vocab_size, vocab_size), dtype='float')
        for word1, word2 in token_pairs:
            i, j = vocab[word1], vocab[word2]
            g[i][j] = 1
            
        # Get Symmeric matrix
        g = self.symmetrize(g)
        
        # Normalize matrix by column
        norm = np.sum(g, axis=0)
        g_norm = np.divide(g, norm, where=norm!=0) # this is ignore the 0 element in norm
        
        return g_norm

    
    def get_keywords(self, number=20):
        """Print top number keywords"""
        return_list = []
        node_weight = OrderedDict(sorted(self.node_weight.items(), key=lambda t: t[1], reverse=True))
        for i, (key, value) in enumerate(node_weight.items()):
            return_list.append(key)
            # print(key + ' - ' + str(value))
            if i > number:
                break
    
        res = self.generate_permutations(return_list, n_gram_range=5)
        res = self.remove_substrings_final(res)
        return res
        
        
    def analyze(self, text, 
                candidate_pos=['NOUN', 'PROPN'], 
                window_size=4, lower=False, stopwords=list()):
        """Main function to analyze text"""
        
        # Set stop words
        self.set_stopwords(stopwords)
        
        # Pare text by spaCy
        doc = nlp(text)
        # print(doc)
        # print("---"*20)
        
        # Filter sentences
        sentences = self.sentence_segment(doc, candidate_pos, lower) # list of list of words
        # print(sentences)
        # print("---"*20)
        
        # Build vocabulary
        vocab = self.get_vocab(sentences)
        
        # Get token_pairs from windows
        token_pairs = self.get_token_pairs(window_size, sentences)
        
        # Get normalized matrix
        g = self.get_matrix(vocab, token_pairs)
        
        # Initionlization for weight(pagerank value)
        pr = np.array([1] * len(vocab))
        
        
        # Iteration
        previous_pr = 0
        for epoch in range(self.steps):
            pr = (1-self.d) + self.d * np.dot(g, pr)
            if abs(previous_pr - sum(pr))  < self.min_diff:
                break
            else:
                previous_pr = sum(pr)

        # Get weight for each node
        node_weight = dict()
        for word, index in vocab.items():
            node_weight[word] = pr[index]
        
        self.node_weight = node_weight

    def generate_permutations(self, words, n_gram_range=6):
        res = []
        for r in range(1, n_gram_range + 1):
            for perm in permutations(words, r):
                phrase = ' '.join(perm)
                if phrase in text:
                    res.append(phrase)
        return list(set(res))

    def remove_substrings_final(self, res):
        final = []
        res = sorted(res)
        for i, kw in enumerate(res):
            # print(f"AT: {kw.upper()}")
            is_final = True
            for j, other_kw in enumerate(res):
                if i == j:
                    # print("SAME")
                    continue
                else:
                    # print(f"Comparing: '{kw}' with '{other_kw}'")
                    if kw in other_kw:
                        # print("BREAK")
                        is_final = False
                        break
                    else:
                        continue
            if is_final:
                # print(f"appended {kw}")
                final.append(kw)
        return final
```

```python
# Let's tokenize the text.
tokens = nltk.word_tokenize(text)
print(tokens[:20])
```

```python
# Now let's tag the tokens with a part-of-speech (POS) tagger.
tagged_tokens = nltk.pos_tag(tokens)
print(tagged_tokens[:20])
```

```python
# Before we build the graph, we need some helper functions.
import re
def is_word(token):
    """
    A token is a "word" if it begins with a letter.
    
    This is for filtering out punctuations and numbers.
    """
    return re.match(r'^[A-Za-z].+', token)

# We only take nouns and adjectives. See the paper for why this is recommended.
ACCEPTED_TAGS = {'NN', 'NNS', 'NNP', 'JJ'}

def is_good_token(tagged_token):
    """
    A tagged token is good if it starts with a letter and the POS tag is
    one of ACCEPTED_TAGS.
    """
    return is_word(tagged_token[0]) and tagged_token[1] in ACCEPTED_TAGS
    
def normalized_token(token):
    """
    Use stemmer to normalize the token.
    """
    return stemmer.stem(token.lower())
```

```python
# Now let's build the graph.
import networkx
graph = networkx.Graph()

# Here, bigrams are "tagged bigrams".
bigrams = nltk.ngrams(tagged_tokens, 2)
for bg in bigrams:
    if all(is_good_token(t) for t in bg):
        normalized = [normalized_token(t[0]) for t in bg]
        graph.add_edge(*normalized)
```

```python
# We can visualize it with matplotlib.
%matplotlib inline
import matplotlib
matplotlib.rcParams['figure.figsize'] = (20.0, 16.0)
networkx.draw_networkx(graph)
```

```python
# Let's do the PageRank.
pagerank = networkx.pagerank(graph)
```

```python
# Then sort the nodes according to the rank.
ranked = sorted(pagerank.items(), key=lambda ns_pair: ns_pair[1], reverse=True)
```

```python
# How many are there?
len(ranked)
```

```python
# We only keep 20% of the top-ranking nodes.
selectivity = 0.20
remove_n = int(len(ranked) * selectivity)
```

```python
# Now remove the nodes we don't need.
for node, _ in ranked[remove_n:]:
    graph.remove_node(node)
```

```python
# Let's see how many are left.
len(graph)
```

```python
# Let's visualize it again.
networkx.draw_networkx(graph)
```

```python
# Now let's recover the key phrases.
edges = graph.edge
phrases = set()

# Using a "sliding window" of size 2, 3, 4:
for n in range(2, 5):
    
    # Get the 2-grams, 3-grams, 4-grams
    for ngram in nltk.ngrams(tokens, n):
        
        # For each n-gram, if all tokens are words, and if the normalized
        # head and tail are found in the graph -- i.e. if both are nodes
        # connected by an edge -- this n-gram is a key phrase.
        if all(is_word(token) for token in ngram):
            head, tail = normalized_token(ngram[0]), normalized_token(ngram[-1])
            
            if head in edges and tail in edges[head]:
                phrase = ' '.join(ngram)
                phrases.add(phrase)
```

```python
# Finally, let's sort the phrases and print them out.
sorted_phrases = sorted(phrases, key=str.lower)
for p in sorted_phrases:
    print(p)
```

```python
import nltk
import re
import networkx as nx
from nltk.stem.snowball import SnowballStemmer

# Assuming text has been defined as in your original code snippet.
# Make sure to download the necessary NLTK resources if you haven't already:
# nltk.download('punkt')
# nltk.download('averaged_perceptron_tagger')

text = df['abstract'][0]

# Tokenization and POS Tagging
tokens = nltk.word_tokenize(text)
tagged_tokens = nltk.pos_tag(tokens)

# Stemmer for normalization
stemmer = SnowballStemmer("english")

def is_word(token):
    return re.match(r'^[A-Za-z].+', token)

ACCEPTED_TAGS = {'NN', 'NNS', 'NNP', 'JJ'}

def is_good_token(tagged_token):
    return is_word(tagged_token[0]) and tagged_token[1] in ACCEPTED_TAGS
    
def normalized_token(token):
    return stemmer.stem(token.lower())

graph = nx.Graph()

bigrams = nltk.ngrams(tagged_tokens, 2)
for bg in bigrams:
    if all(is_good_token(t) for t in bg):
        normalized = [normalized_token(t[0]) for t in bg]
        graph.add_edge(*normalized)

pagerank = nx.pagerank(graph)
ranked = sorted(pagerank.items(), key=lambda ns_pair: ns_pair[1], reverse=True)
selectivity = 0.20
remove_n = int(len(ranked) * selectivity)

for node, _ in ranked[remove_n:]:
    graph.remove_node(node)

phrases = set()
for n in range(2, 5):
    for ngram in nltk.ngrams(tokens, n):
        if all(is_word(token) for token in ngram):
            normalized_ngram = [normalized_token(token) for token in ngram]
            # Check all bigrams within the n-gram are present in the graph
            present_in_graph = all(graph.has_edge(normalized_ngram[i], normalized_ngram[i+1]) 
                                   for i in range(len(normalized_ngram)-1))
            if present_in_graph:
                phrase = ' '.join(ngram)
                phrases.add(phrase)

sorted_phrases = sorted(phrases, key=str.lower)
for p in sorted_phrases:
    print(p)
```

```python
import nltk
import re
import networkx as nx
from nltk.stem.snowball import SnowballStemmer

# Ensure you've downloaded the necessary NLTK resources
# nltk.download('punkt')
# nltk.download('averaged_perceptron_tagger')

def extract_keyphrases(text, max_n=5):
    """
    Extract keyphrases from the given text considering n-grams from 1 to max_n.
    
    Parameters:
    - text (str): The text from which to extract keyphrases.
    - max_n (int): The maximum length of n-grams to consider.
    
    Returns:
    - list: A list of extracted keyphrases.
    """
    
    # Tokenize and tag the text
    tokens = nltk.word_tokenize(text)
    tagged_tokens = nltk.pos_tag(tokens)
    
    # Initialize the stemmer
    stemmer = SnowballStemmer("english")
    
    def is_word(token):
        return re.match(r'^[A-Za-z].+', token)
    
    ACCEPTED_TAGS = {'NN', 'NNS', 'NNP', 'JJ'}
    
    def is_good_token(tagged_token):
        return is_word(tagged_token[0]) and tagged_token[1] in ACCEPTED_TAGS
    
    def normalized_token(token):
        return stemmer.stem(token.lower())
    
    # Build the graph
    graph = nx.Graph()
    for n in range(1, max_n + 1):
        ngrams = nltk.ngrams(tagged_tokens, n)
        for ng in ngrams:
            if all(is_good_token(t) for t in ng):
                normalized = [normalized_token(t[0]) for t in ng]
                for start, end in zip(normalized[:-1], normalized[1:]):
                    graph.add_edge(start, end)
    
    # Calculate PageRank
    pagerank = nx.pagerank(graph)
    ranked = sorted(pagerank.items(), key=lambda ns_pair: ns_pair[1], reverse=True)
    
    # Select the top keyphrases
    selectivity = 0.20
    selected_n = int(len(ranked) * (1 - selectivity))
    selected_nodes = set([node for node, _ in ranked[:selected_n]])
    
    phrases = set()
    for n in range(1, max_n + 1):
        for ngram in nltk.ngrams(tokens, n):
            if all(is_word(token) for token in ngram):
                normalized_ngram = [normalized_token(token) for token in ngram]
                if all(node in selected_nodes for node in normalized_ngram):
                    phrases.add(' '.join(ngram))
    
    # Return sorted phrases
    return sorted(phrases, key=str.lower)

# Example usage
text = df['abstract'][0]
keyphrases = extract_keyphrases(text, max_n=5)
for p in keyphrases:
    print(p)
```
